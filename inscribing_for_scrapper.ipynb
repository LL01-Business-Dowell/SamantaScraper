1. PRE-INSERTION

Calculation and transformation of location coordinates using inscribing function 

```python
import numpy as np
from math import sqrt
import pandas as pd
import itertools
import ast
import json
import re
from openpyxl import load_workbook
from pymongo import MongoClient, GEOSPHERE
import requests
from geopy.distance import geodesic
import folium
```

```python
input_radius = float(input("Please enter a radius for the in-circles: "))
input_length = int(input("Please enter the length of the canvas: "))
input_width = int(input("Please enter the width of the canvas: "))
```

```python
radius = input_radius
length = input_length
width = input_width

hyp = 2*radius
height = radius*sqrt(3)
```

```python
# iterations for x coordinates

def seq1(start,length,step):
        y = []
        x = []
        iterations = int(10*length)
        i = 0
        count = start


        while i < iterations:
            y.append(count)
            if i>0:
                x.append(-count)
            if i != iterations:
                count += step
            i += 1
            if count>((length/2)):
                break

        w =  list(reversed(x))
        w.extend(y)
        
        x_num = len(w)
        print("Total no. of X-coordinates: ",x_num)

        return w

a = seq1(start = 0, length = length, step = radius)
a
```

```python
#iterations for y coordinates
    
def seq2(start, width, step):
    y = []
    x = []
    iterations = 10*width
    i = 0
    count = start

    while i < iterations:
        y.append(count)
        if i>0:
            x.append(-count)
        if i != iterations:
            count += step
        i += 1
        if count>(width/2):
            break

    w = list(reversed(y))        
    w.extend(x)

    y_num = len(w)
    print("Total no. of Y-coordinates: ",y_num)

    return w

b = seq2(start = 0, width = width, step = radius)
b
```

```python
# preparing the dataframe of the coordinates
    
df = pd.DataFrame(columns=a,index=b)
df.replace(to_replace=np.NaN,value="",inplace=True)
df1 = df.copy() #changes in df1 will not effect in df

# placing the center circle (0,0) first in the dataframe
df1.at[0, 0] = (0, 0)

# placing the values as (x,y) coordinates
for i in b:
    for j in a:
        df.at[i,j]=[i,j]

# getting the odd and even indexing value of index seperating b into even indexing and odd indexing of b
odd_ind = [b[i] for i in range(0, len(b)) if i % 2 != 0] 
even_ind = [b[i] for i in range(0, len(b)) if i % 2 == 0]

# getting the odd and even columns value of index seperating a into even columns and odd columns  of a
odd_col = [a[i] for i in range(0, len(a)) if i % 2 != 0] 
even_col = [a[i] for i in range(0, len(a)) if i % 2 == 0]

# for placing the even columns
for i in even_ind:
    for j in even_col:
        df1.at[i,j]=(j,i)

# for placing the even columns around (0,0)
for i in even_ind:
    for j in even_col:
        if (i, j) != (0, 0):
            df1.at[i, j] = (j, i)

# for placing the odd columns around (0,0)
for i in odd_ind:
    for j in odd_col:
        if (i, j) != (0, 0):
            df1.at[i, j] = (j, i)
```

```python
# transforming the cartesian coordinates to geocoordinates
"""
1 KM = 0.00899321605918700 degrees
Multiply (x,y) by the scaling factor t = 0.00899321605918700
Add the latitude and longitude of the desired location to the scaled (x,y) to shift the origin to the desired location

"""
t = 0.00899321605918700 
lat, lon = 12.953945600000000, 77.466127900000000   #Bangalore, India

def scale_coordinates(x):
    if pd.isna(x) or x == '':
        return x 
    try:
        coord = ast.literal_eval(x) if isinstance(x, str) else x
        return tuple((t*coord[1], t*coord[0]))
    except Exception as e:
        print(f"Skipping cell '{x}': {e}")
        return np.nan

df_scaled = df1.applymap(scale_coordinates)


"""
These transformed coordinates will be saved into the master db using datacube
"""

df_scaled.head(150)
```

2. INSERTION

Creating and populating the master db

```python
""""
Connect the datacube APIs for creating the db and inserting the data
"""
```

```python
def store_coordinates(file_path, mongo_uri='mongodb://localhost:27017/'):
    """
    Reads coordinates from Excel, stores in MongoDB with correct (lat, lon) order,
    and creates optimized index collections.
    """
    client = MongoClient(mongo_uri)
    db_raw = client['raw_coordinates_db']
    db_geo = client['geocoordinates_db']
    
    # Create unified index collection
    index_db = client['coordinate_index_db']
    lat_index_coll = index_db['latitude_index']
    lat_index_coll.create_index("latitude")
    
    # Track processed latitudes
    processed_lats = set()
    
    wb = load_workbook(filename=file_path, read_only=True)
    sheet = wb.active
    
    for row_idx, row in enumerate(sheet.iter_rows(min_row=2)):
        for col_idx, cell in enumerate(row[1:], start=1):
            value = cell.value
            if not value or not isinstance(value, str):
                continue
                
            match = re.match(r'\(([^,]+),\s*([^)]+)\)', value.strip())
            if not match:
                continue
                
            # Extract and swap values: first is latitude, second is longitude
            lat_str, lon_str = match.groups()
            try:
                lat_val = float(lat_str.strip())
                lon_val = float(lon_str.strip())
            except ValueError:
                continue
                
            # Create collection name from latitude
            collection_name = f"lat_{lat_val:.6f}"
            
            # Insert into raw database
            raw_coll = db_raw[collection_name]
            raw_coll.insert_one({
                'latitude': lat_val,
                'longitude': lon_val
            })
            
            # Insert into GeoJSON database
            geo_coll = db_geo[collection_name]
            geo_coll.insert_one({
                'location': {
                    'type': 'Point',
                    'coordinates': [lon_val, lat_val]  # GeoJSON: [lon, lat]
                }
            })
            
            # Create geospatial index for new collections
            if lat_val not in processed_lats:
                geo_coll.create_index([('location', GEOSPHERE)])
                # Add to unified index
                lat_index_coll.insert_one({
                    'latitude': lat_val,
                    'collection': collection_name
                })
                processed_lats.add(lat_val)
    
    wb.close()
    client.close()
```

```python
store_coordinates(f"scaled_coordinates_{length}_{width}.xlsx")
```

3. PRE-FETCHING

Calculating the boundary points of the users canvas and running the fetch query to filter the coordinates from the db

```python
""""
user inputs: distance from the center (d), latitude (lat), longitude (lon)

"""

# determine the boundary points of the users canvas

d = 10
lat = 0.000000000000000
lon = 0.000000000000000
t = 0.00899321605918700 

top_left = tuple((lat + d*t, lon - d*t))
top_right = tuple((lat + d*t, lon + d*t))
bottom_left = tuple((lat - d*t, lon - d*t))
bottom_right = tuple((lat - d*t, lon + d*t))

print(f"Top Left: {top_left}, Top Right: {top_right}, Bottom Left: {bottom_left}, Bottom Right: {bottom_right}")

top_mid = tuple((lat + d*t, lon))
bottom_mid = tuple((lat - d*t, lon))
left_mid = tuple((lat, lon - d*t))
right_mid = tuple((lat, lon + d*t))

print(f"Top Mid: {top_mid}, Bottom Mid: {bottom_mid}, Left Mid: {left_mid}, Right Mid: {right_mid}")
```

4. FETCHING

```python
"""""
Boundary points of the users canvas will be passed in the fecth query. Datacube will filter and return the geocoordinates falling within this boundary.
"""
```

```python
url = "http://localhost:8000/api/geo-query/"

payload = {
    "top_left":top_left,
    "top_right":top_right,
    "bottom_left":bottom_left,
    "bottom_right":bottom_right
}

request = requests.post(url = url, json=payload)
response = request.json()
response
```
